1. What is a RESTful API?
- A RESTful API is a web API that follows the principles of REST (Representational State Transfer) to let clients and servers communicate over HTTP in a simple, standardized way. In practice, you use URLs to represent resources and HTTP methods like GET, POST, PUT, and DELETE to work with those resource .
2. What is Flask, and why is it popular for building APIs?
- Flask is a lightweight web framework in Python that makes it easy to build web applications and APIs. It is called a “microframework” because it provides the essentials (routing, request/response handling, templates) without enforcing a specific database, authentication system, or project structure.
3. What are HTTP methods used in RESTful APIs?
- GET
    - Used to fetch/read data from the server without changing anything on the server .
- POST
    - Used to create new data (a new resource) on the server.
- PUT
    - Used to completely update or replace an existing resource on the server.
- DELETE
    - Used to remove an existing resource from the server.

4. What is the purpose of the @app.route() decorator in Flask?
- The @app.route() decorator in Flask is used to bind a URL endpoint to a Python function. When a user visits that URL, Flask executes the associated function and returns its result as an HTTP response .  
5. What is the role of Flask-SQLAlchemy?
- Flask-SQLAlchemy is an extension for Flask that simplifies database operations by providing an ORM (Object-Relational Mapper). It allows you to interact with databases using Python classes and objects, making it easier to perform CRUD operations without writing raw SQL .

8. How do you handle POST requests in Flask?
- from flask import request

    data = request.form['key']
    return f'Received: {data}'
